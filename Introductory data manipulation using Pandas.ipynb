```python
#Creating a dataframe which is a multi-dimensional table that is size-mutable and has multiple rows & column names.
# We have not provided indexing information so index is defaulted to range index (0,1,2,3 etc)

df = pd.DataFrame({
    'Population': [35.467, 63.951, 80.94 , 60.665, 127.061, 64.511, 318.523],
    'GDP': [1785387,2833687,3874437,2167744,4602367,2950039,17348075],
    'Surface Area': [ 9984670,640679,357114,301336,377930,242495,9525067],
    'HDI': [0.913,0.888,0.916,0.873,0.891,0.907,0.915 ],
    'Continent': ['America','Europe','Europe','Europe','Asia','Europe','America']})

df
```

```python
#We have added in indexing information so index has changed from range index.
df.index = ['Canada','France','Germany','Italy','Japan','United Kingdom','United States',]
```

# Conditional selection (boolean arrays)

```python
#Selects all countries and says True or False as to whether they have a population which is less than 70
df['Population'] > 70 
```

```python
# Displays all countries wich have a population that is greater than 70.
#We use loc rather than iloc, because loc is an explicit index that can access rows or column specified by a particular label or to access a boolean array.

Population_filter = df.loc[df['Population'] > 70] 
```

```python
Population_filter
```

```python
#Showing only the Population and GDP columns for countries where Population is greater than 70.
df.loc[df['Population'] > 70, ['Population', 'GDP']] 
```
